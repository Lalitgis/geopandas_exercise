```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('Scraped.csv')
data.head()
```

```python
# remove few columne and rename the name in to districts

data.drop(columns = ['Pop_2001', 'Pop_2011'], inplace=True)
```

```python
data.head()
```

```python
data.rename(columns = {'Name': 'Districts', 'Pop_2021': 'Population'}, inplace=True)
```

```python
data.info()
```

```python
# Import the shapefile

district = gpd.read_file('./Shapefiles/districts.shp')
```

```python
district.plot()
```

```python
district.head()
```

```python
# Drop unnecessary columns

district.drop(columns = ['FIRST_STAT', 'FIRST_DIST'], inplace=True)
```

```python
district.head(40)
```

```python
district.rename(columns= {'DISTRICT': 'Districts'}, inplace=True)
```

```python
district
```

```python
data['Districts'] = data['Districts'].str.upper()
```

```python
data.head()
```

```python
# Change the coordinate system

crs = district.crs
crs
```

```python
# Change the crs into utm/projected coordinate system

district.to_crs(epsg=32645, inplace=True)
```

```python
for index, row in district['Districts'].iteritems():
    if row in data['Districts'].tolist():
        pass
    else:
        print('The district', row, 'is NOT in the population data list')
```

```python
data.replace('CHITWAN', 'CHITAWAN', inplace=True)
data.replace('DANG DEUKHURI', 'DANG',inplace=True)
data.replace('DHANUSA','DHANUSHA',inplace=True)
data.replace('KAVREPALANCHOK','KABHREPALANCHOK',inplace=True)
data.replace('KANCHANPUR','KANCHANPUR',inplace=True)
data.replace('KAPILVASTU','KAPILBASTU',inplace=True)
data.replace('MAKWANPUR','MAKAWANPUR',inplace=True)
data.replace('NAWALPARASI EAST','NAWALPARASI_E',inplace=True)
data.replace('NAWALPARASI WEST','NAWALPARASI_W',inplace=True)
data.replace('EAST RUKUM','RUKUM_E',inplace=True)
data.replace('WEST RUKUM','RUKUM_W',inplace=True)
data.replace('SINDHUPALCHOWK','SINDHUPALCHOK',inplace=True)
data.replace('TANAHUN','TANAHU',inplace=True)
data.replace('TEHRATHUM','TERHATHUM',inplace=True)
```

```python
for index, row in district['Districts'].iteritems():
    if row in data['Districts'].tolist():
        pass
    else:
        print('The district', row, 'is NOT in the population data list')
```

```python
# calculate area
district.head(2)
district['area'] = district.area/1000000
# area is in sq.km
```

```python
district.head(2)
```

```python
# Now we join the two attributes ie popualtion data and shape file
district = district.merge(data, on = 'Districts')
```

```python
district.head(2)
```

```python
# Now we calcualte the popualation density

district['pop_den (people/sq.km)'] = district['Population']/district['area']
```

```python
# Now we plot the map
ax, fig = figsize=(20,15)
ax = district.plot(column = 'pop_den (people/sq.km)', cmap= 'Spectral', legend = True,
                   legend_kwds={"label": "Population density of Nepal 2021", "orientation": "horizontal"},figsize =(10,10))
ax.set_title("Population Density Map of Nepal")
```
