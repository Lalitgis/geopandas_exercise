```python
import geopandas as gpd
```

```python
data = gpd.read_file('Shapefiles/Districts.shp')
```

```python
data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(figsize =(10,10))
data.plot(ax =ax, column="Area", cmap = 'Blues')
ax.axis("off")
ax.set_title("Map of Nepal showing area of different districts", fontdict={
    'fontsize': 20, 'fontweight': 5})


```

```python
hq = gpd.read_file('Shapefiles/DIS_HQ.shp')
```

```python
hq.head()
```

```python
fig, ax = plt.subplots(figsize =(10,10))
data.plot(ax =ax, column="Area", cmap = 'Blues')
hq.plot(ax=ax, column = "VIL_NAME")
ax.axis("off")
ax.set_title("Map of Nepal showing area of different districts", fontdict={
    'fontsize': 20, 'fontweight': 5})
```
